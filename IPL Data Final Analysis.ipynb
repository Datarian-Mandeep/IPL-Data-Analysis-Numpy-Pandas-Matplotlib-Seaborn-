# Data Understanding & Data Cleaning

```python
a.info()
a.isnull().sum(axis=0)
```

```python
a[a.duplicated()]                # To check duplicate record
a.drop_duplicates()              # To drop duplicate records
```

```python
a[a.city.isna()]         # Records where city is NaN/Null
a[(a.venue=="Dubai International Cricket Stadium") | (a.venue=="Sharjah Cricket Stadium")]
a.groupby(['city'])['venue'].value_counts()
```

```python
a.loc[a.venue=="Dubai International Cricket Stadium","city"]="Dubai"      # filling city null values
a.loc[a.venue=="Sharjah Cricket Stadium","city"]="Sharjah"
```

```python
a1=a.loc[a.player_of_match.isna(),'id']     # dynamically removed null values in player_of_match
a1.reset_index()
a2 = list(a1.index)
a2
a.drop(a2,axis=0,inplace=True)
```

```python
a.loc[a.result=="tie",["winner"]]
a.loc[a.result=="tie",["winner"]]="No_winner"    # tie result means no_winner
```

```python
a.result_margin.fillna(0,inplace=True)         # filling result_margin null values
```

```python
a.method.fillna("no_method",inplace=True)      # filling method null values
```

```python
a['id']=a['id'].astype("str")                              # Type casting of various columns
a['date']=a['date'].astype('datetime64[ns]')
a['result_margin']=a['result_margin'].astype("int")
a['year']=pd.DatetimeIndex(a['date']).year                   # how to fetch year from date column
a['year']
```

```python
a26=a.winner.value_counts()                  # how many times each team won the matches
a26=a26.reset_index()
a26.columns=['winner_team',"Matches_won"]
a26=a26.sort_values(by="winner_team")
a26=a26.drop(10)
a26
a27=list(a26.Matches_won)

a23=a.team1.value_counts()
a24=a.team2.value_counts()
frames=[a23,a24]
c1=pd.concat(frames,axis=0)                      # Total no of Matches each team played
c1=pd.DataFrame(c1)
c1=c1.reset_index()
c1.columns=['Teams',"Total_Matches_Played"]
a25=c1.groupby(['Teams'],as_index=False)['Total_Matches_Played'].sum().sort_values(by="Teams",ascending=True)
a25['Matches_won']=a27
a25     # Final Dataset
# a25=a25.reset_index()

plt.figure(figsize=(8,6))
plt.barh(a25.Teams, a25.Matches_won, color='magenta')
plt.barh(a25.Teams, a25.Total_Matches_Played, left=a25.Matches_won, color='pink')
plt.ylabel("Teams",color="b",fontsize=15)
plt.legend(['Matches_won','Total_Matches_Played'])
plt.show()

a28=a25[['Teams','Total_Matches_Played','Matches_won']]
a28.plot.bar(x="Teams",width=0.6)
```
